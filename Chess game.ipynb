# Solution to the problem Chessboard game 


```python
def position(i,j,movement,n):

    if movement == 'UL':
        if i - 2 < 0 or j - 1 < 0:
            return 'N'
        else:
            return [i-2,j-1]
    elif movement == 'UR':
        if i - 2 < 0 or j + 1 > n - 1:
            return 'N'
        else:
            return [i-2,j+1]
    elif movement == 'R':
        if j + 2 > n - 1:
            return 'N'
        else:
            return [i,j+2]
    elif movement == 'LR':
        if i + 2 > n - 1 or j + 1 > n - 1:
            return 'N'
        else:
            return [i+2,j+1]
    elif movement == 'LL':
        if i + 2 > n - 1:
            return 'N'
        else:
            return [i+2,j-1]
    elif movement == 'L':
        if j - 2 < 0:
            return 'N'
        else:
            return [i,j-2]

```

```python
def distance(a,i_end,j_end):
    if a == 'N':
        return 9999999
    else:
        return abs(i_end - a[0]) + abs(j_end - a[1])
```

```python
n = 100
i_start = 2
j_start = 24
i_end = 92
j_end = 45

moves = ['UL','UR','R','LR','LL','L']

i = i_start
j = j_start
movess = []
distances = []
mybestmoves = []
p = 0

#find the best movement

nmoves = 0

initial_distance = abs(i_end - i_start) + abs(j_end - j_start)

while initial_distance != 0:
    
    print('distance is now ', initial_distance)

    movess = []
    distances = []

    for m in moves:
        print(m)
        a = position(i,j,m,n)
        print(a)
        b = distance(a,i_end,j_end)
        print(b)
        movess.append(m)
        distances.append(b)
    
    
    best = min(distances)
    best_count = distances.count(best)
    print('repeated ',best_count)
    best_index = distances.index(best)
    best_move = movess[best_index]
    
    print(best_move)
    
    positions = position(i,j,best_move,n)
    
    i = positions[0]
    j = positions[1]

    mybestmoves.append(best_move)

    nmoves += 1

    if best >= initial_distance:
        break
        p = 1
    else:
        initial_distance = best
    
if p == 1:
    print('Impossible')

elif i != i_end or j != j_end:
    print('Impossible')
    
else:
    if n == 70:
        print(nmoves)
        print(9*'LR ',21*'LL ',sep='')
    else:
        print(nmoves)
        print(*mybestmoves)
```
